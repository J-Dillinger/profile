```python
import pandas as pd
import re
import numpy as np
```

```python
def striphtml(data):
    p = re.compile(r'<.*?>')
    return p.sub('', data)
```

```python
def stripnum(data):
    return re.findall(r'\d+', data)
```

```python
def striphtml7(data):
    p = re.sub("\\\\n", "", data)
    return p
```

```python
def bc(data):
    p = re.findall(r'^(.+?),',data)
    return p
```

```python
rdf = pd.read_csv('/Users/yonilevine/Desktop/ratingssel3f2.csv')
```

```python
rdf.drop('Unnamed: 0',axis =1 ,inplace=True)
```

```python
rdf['text'] = rdf['text'].apply(striphtml)
rdf['text'] = rdf['text'].apply(striphtml7)
rdf['text'] = [item.strip('\[|\]|\'') for item in rdf.text]
```

```python
rdf.scorew= rdf.scorew.apply(stripnum)

scores = []
for row in rdf.scorew:
    x=map(int, row)
    scores.append(x)
    
rdf['scorew']=scores    

rdf['scorew']=[sum(item) for item in rdf.scorew]
```

```python
rdf
```

```python
rdf['scoremb']=rdf.scoremb.apply(stripnum)
rdf['scoremb2']=rdf.scoremb2.apply(stripnum)
rdf['scoremb3']=rdf.scoremb3.apply(stripnum)
```

```python

smb=[]
for item in rdf.scoremb:
    if len(item)==2:
        smb.append(item)
    elif len(item)>2:
        smb.append(item[-2:])     
    else:
        smb.append([])
        
rdf['scoremb']=smb      

smb2=[]
for item in rdf.scoremb2:
    if len(item)==2:
        smb2.append(item)
    elif len(item)>2:
        smb2.append(item[-2:])     
    else:
        smb2.append([])
        
rdf['scoremb2']=smb2      
        
        
        
smb3=[]
for item in rdf.scoremb3:
    if len(item)==2:
        smb3.append(item)
    elif len(item)>2:
        smb3.append(item[-2:])     
    else:
        smb3.append([])
        
        
rdf['scoremb3']=smb3  
```

```python
rdf['scorew']=rdf['scorew']/rdf['numw']
```

```python
r=rdf['scoremb']+rdf['scoremb2']+rdf['scoremb3']
```

```python
score=[]
num=[]
for row in r:
    if len(row)>0:
        score.append(row[0])
        num.append(row[1])
    else:
        score.append('0')
        num.append('0')
```

```python
rdf['scorembf']=pd.to_numeric(score,errors='coerce')
rdf['nummb']=pd.to_numeric(num,errors='coerce')
```

```python
rdf.drop("scoremb scoremb2 scoremb3".split(),axis=1,inplace=True)
```

```python
rdf.scorew.fillna(0,inplace=True)
```

```python
rdf['scorew']=rdf.scorew.astype(int)
```
