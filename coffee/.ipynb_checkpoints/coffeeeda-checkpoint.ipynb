```python
df=pd.concat([df,bd,rld],axis=1)
```

```python
# Creating dummy variables for the country data
df.country = df.country.apply(str.split)
for (i,list_) in enumerate(df.country):
    for item in list_:
        if item not in df.country:
            df[item] = 0
        df.set_value(col=item,index=i,value=1)
```

```python
df.process = df.process.apply(str.split)
for (i,list_) in enumerate(df.process):
    for item in list_:
        if item not in df.process:
            df[item] = 0
        df.set_value(col=item,index=i,value=1)
```

```python
great=[]
for item in df.scoref:
    if item >80:
        great.append(1)
    else:
        great.append(0)
df['great']=great
```

```python
for item in df.columns:
    print item,
```

```python
mdf = df.drop("index url title roaster roastlevel blend description country process text scorew numw scorembf nummb scoref".split(),axis=1)
```

```python

X = mdf[mdf.columns[:-1]]
y= mdf.great
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=.3,stratify=great)
```

```python
param_grid={'n_estimators':[25,50,75],
           'max_features':[1,2,3,4,5],
           'max_depth':[2,4,6,8]
           }
rf=RandomForestClassifier()
grid=GridSearchCV(estimator=rf, param_grid=param_grid, cv=5)
grid.fit(X_train,y_train)
print grid.best_params_
print grid.best_score_   
```

```python
best=grid.best_estimator_
model=best.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
all(model.feature_importances_ == np.mean([tree.feature_importances_ for tree in model.estimators_], axis=0))

importances = model.feature_importances_
# calculate the standard deviation of feature importances by looping over the trees in the random forest
# 

std = np.std([tree.feature_importances_ for tree in model.estimators_], axis=0)

indices = np.argsort(importances)[::-1]
feature_names = X_train.columns

# Plot the feature importances of the forest
plt.figure(figsize=(20,10))
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(X.shape[1]), feature_names[indices], rotation=90)
plt.xlim([-1, X.shape[1]])
plt.show()
```

```python
param_grid={'penalty':['l1','l2'],
           'C':[1,2,3,4,5,6,7,8],
           }
lor=LogisticRegression()
grid2=GridSearchCV(estimator=lor, param_grid=param_grid, cv=5)
grid2.fit(X_train,y_train)
print grid2.best_params_
print grid2.best_score_   
print grid2.best_estimator_
best2=grid2.best_estimator_
```

```python
modellr=best2.fit(X_train,y_train)
modellr.score(X_test,y_test)
```
